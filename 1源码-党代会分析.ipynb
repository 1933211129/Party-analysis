```python
import os

def merge_txt(path, result):
    # 获取目标文件夹的路径
    # path = "E:/paper1_历届党代会报告分析/历届党代会报告/P1/"
    # 获取当前文件夹中的文件名称列表
    filenames = os.listdir(path)
    # 打开当前目录下的result.txt文件，如果没有则创建
    file = open(result, 'w+', encoding="utf-8")
    # 向文件中写入字符
    
    # 先遍历文件名
    for filename in filenames:
        filepath = path + '/'
        filepath = filepath + filename
        # 遍历单个文件，读取行数
        for line in open(filepath, encoding="utf-8"):
            file.writelines(line)
        file.write('\n')
    # 关闭文件
    file.close()
```

## 剔除停用词并保存为npy

```python
def npy(name, npy):
    import numpy as np

    with open('stopwords.txt','r') as f:
        stop_words = f.read()
    words = np.load(npy).tolist()
    fliter_list = [w for w in words if not w in stop_words]

    np.save(name, fliter_list)
    return fliter_list
```

# TF-IDF词云图

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# score字典
score = []
plt.rcParams['font.family'] = 'SimHei'
corpus = [' '.join(p1), ' '.join(p2), ' '.join(p3)]

vectorizer = TfidfVectorizer(use_idf=True)

# 对三个阶段的文本内容进行TF-IDF向量化
tfidf_matrix = vectorizer.fit_transform(corpus)

# 为每个阶段生成词云图
for i, tfidf in enumerate(tfidf_matrix):
    # 使用TF-IDF值为每个词汇进行打分
    scores = {word: score for word, score in zip(vectorizer.get_feature_names(), tfidf.toarray()[0])}
    score.append(scores)
    
    wc = WordCloud(font_path='simhei.ttf', width=800, height=600, background_color='white', max_words=250)
    
    wordcloud = wc.generate_from_frequencies(scores)
    
    plt.figure(figsize=(8, 6))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.title('阶段 {}'.format(i + 1))
    plt.show()
```

```python
def custom_dispersion_plot(text, words, stage_boundaries=[86859, 217909]):
    import matplotlib.pyplot as plt
    from nltk.text import Text

    plt.rcParams['font.family'] = 'SimHei'

    text = Text(text)
    words.reverse()
    
    points = [(x,y) for x in range(len(text)) 
              for y in range(len(words)) if text[x] == words[y]]

    if points:
        x, y = zip(*points)
    else:
        x = y = ()

    plt.plot(x, y, "b|", scalex=.1)
    plt.yticks(list(range(len(words))), words, color="b")
    plt.ylim(-1, len(words))
    plt.title("词汇分散图")
    plt.xlabel("Word Offset")


    stages = ['第一阶段', '第二阶段', '第三阶段']
    for i, boundary in enumerate(stage_boundaries):
        plt.axvline(x=boundary, color='r', linestyle='--')
        plt.text(boundary, -1, stages[i], fontsize=12, ha='right')
        

    plt.text(len(text), -1, stages[-1], fontsize=12, ha='right')


    plt.xticks(stage_boundaries + [len(text)])

    plt.show()
```

## 文章分句

```python
import re

def split_sentences(text):
    # 更新标点符号列表，去除括号和引号
    sentences = re.split('(。|！|\!|\.|？|\?|；|：|……|—)', text)
    sentences = [sent.strip() for sent in sentences if sent.strip() not in ['。', '！', '!', '.', '？', '?', '；', '：', '……', '—', ' ']]
    return sentences

# with open('./raw_corpus/P1/1921一大.txt', 'r', encoding='utf-8') as f:
#     text = f.read()
    
# sentences = split_sentences(text)
```

## 对每一个阶段分句再分词，不去停用词，保存为npy

```python
def npy_vec(file, name):
    import jieba
    import numpy as np

    # dict_path = r'E:\1新党代会分析\dict\dict.txt'
    # jieba.load_userdict(dict_path)
    with open(file, 'r', encoding='utf-8') as f:
        text = f.read()
    sentences = split_sentences(text)
    sentences_words = [jieba.lcut(sent) for sent in sentences]

    np.save(name, np.array(sentences_words, dtype=object))
    return sentences_words
```

## get_month_corpus

```python
def get_month_corpus(corpus):
    from gensim import corpora
    dictionary = corpora.Dictionary(corpus)
    # dictionary.save('{}.dict'.format(time))  # 把字典存储下来，可以在以后直接导入

    dictionary.filter_extremes(no_below=10, no_above=0.5)  # 删掉只在不超过20个文本中出现过的词，删掉在50%及以上的文本都出现了的词
    # dictionary.filter_tokens(['中国','一年'])  # 这个函数可以直接删除指定的词
    # dictionary.compactify()  # 去掉因删除词汇而出现的空白

    corpus = [dictionary.doc2bow(s) for s in corpus]
    corpora.MmCorpus.serialize('./LDA_model/corpus_bow_p3.mm', corpus)  # 存储语料库
    return dictionary,corpus
```
